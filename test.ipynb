When you get to a quoted literal like `'A'`, that's called a "terminal symbol." It means you're done rewriting that symbol. 

The other symbols without quotes are called "non-terminal." So you keep expanding the non-terminal symbols, until you're left with only terminal symbols.

So, we saw the derivation, step by step. We learned some formal grammar terminology: Terminal and Non-terminal symbols. 

And here's what "context-free grammar" means:
- In a context-free rule, the left side is always a single non-terminal symbol.
- The right side is a finite sequence of terminals and/or non-terminals.
- Because the left side of the rule is a single non-terminal symbol, the rule does not depend on context from other symbols.
